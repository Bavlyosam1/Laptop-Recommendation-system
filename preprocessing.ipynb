```python
import pandas as pd

df = pd.read_csv(r'C:\Users\MK\Downloads\scraping_dl\scraping_dl\laptops_data_clean.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill missing values with median
df['storage_gb'] = df['storage_gb'].fillna(df['storage_gb'].median())
df['screen_size_inch'] = df['screen_size_inch'].fillna(df['screen_size_inch'].median())

df.isnull().sum()  
```

```python
df.duplicated().sum() 
```

```python
print(df.duplicated().sum()) 
```

```python
# Detect outliers in storage_gb
print(df['storage_gb'].describe())
print(df[df['storage_gb'] > 10000])  
```

```python
# Remove outliers
df = df[df['storage_gb'] <= 10000]
```

```python
# Replace unknown values
df['cpu_family'] = df['cpu_family'].replace('Unknown', df['cpu_family'].mode()[0])
df['storage_type'] = df['storage_type'].replace('Unknown', 'SSD')
```

```python
print(df['cpu_family'].value_counts())
print(df['storage_type'].value_counts())
```

```python
# Encode categorical columns to numbers
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ['storage_type', 'cpu_family', 'cpu_generation', 'cpu_model', 'gpu']:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
# Save as CSV
df.to_csv(r'C:\Users\MK\Downloads\scraping_dl\scraping_dl\laptops_preprocessed.csv', index=False)
```
